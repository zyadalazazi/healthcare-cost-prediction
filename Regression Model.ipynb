```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
health_data = pd.read_csv(r'C:\Users\MEPI\Desktop\LAU\Springg 2021\CSC491\Week 4\insurance.csv')

df_health = pd.DataFrame(health_data)
df_health
```

```python
df_health.nunique()
```

```python
df_health.info()
```

```python
label_encoder = LabelEncoder()

df_health["sex"] = label_encoder.fit_transform(df_health["sex"])## female is encoded as 0 and male is encoded as 1
df_health["smoker"] = label_encoder.fit_transform(df_health["smoker"])## yes is encoded as 1 and no is encoded as 0
df_health["region"] = label_encoder.fit_transform(df_health["region"])## southwest as 3, southeast as 2, northwest as 1 and northeast as 0 

df_health
```

```python
df_health.describe()
```

```python
charges = df_health["charges"]

charges
```

```python
df_health.pop('charges')

df_health
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_health, charges, test_size = 0.2, random_state = 100)

x_incls_cons = sm.add_constant(x_train)
model = sm.OLS(y_train, x_incls_cons)
results = model.fit()

pd.DataFrame({'coef':results.params, 'p-value':round(results.pvalues, 9)})
```

```python
x_train = x_train.drop('sex', axis = 1)
x_train = x_train.drop('region', axis = 1)

x_test = x_test.drop('sex', axis = 1)
x_test = x_test.drop('region', axis = 1)

final_model = make_pipeline(StandardScaler(), LinearRegression())
final_model.fit(x_train, y_train)
final_model.score(x_test, y_test)
```
